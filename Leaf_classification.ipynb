```python
import cv2
path = 'C:\\Users\\Sankalp Shukla\\Downloads\\ample/'
im_size = 227

images = []
labels = []

for i in room_types:
    print(i)
    data_path = path + str(i)  # entered in 1st folder and then 2nd folder
    filenames = [i for i in os.listdir(data_path) ]
    # print(filenames)  # will get the names of all images
    for f in filenames:
        img = cv2.imread(data_path + '/' + f)  # reading that image as array
        #print(img)  # will get the image as an array
        img = cv2.resize(img, (im_size, im_size))
        images.append(img)
        labels.append(i)
        if len(os.listdir(data_path))<123:
            im = cv2.flip(img,0)
            images.append(im)
            labels.append(i)
            if len(os.listdir(data_path))<60:
                im = cv2.flip(img,1)
                images.append(img)
                labels.append(i)
                if len(os.listdir(data_path))<30:
                    image = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
                    images.append(img)
                    labels.append(i)
```

```python
IMG_HEIGHT = 227
IMG_WIDTH = 227
IMG_CHANNELS = 3
```

```python
## Architecture
inputs = tf.keras.layers.Input((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS))

c1 = tf.keras.layers.Conv2D(filters = 96,kernel_size=(11,11),strides = 4, activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
c1 = tf.keras.layers.BatchNormalization()(c1)
p1 = tf.keras.layers.MaxPool2D(pool_size=(3,3),strides=2)(c1)

c2 = tf.keras.layers.Conv2D(filters = 256,kernel_size=(5,5),strides = 1, activation='relu', kernel_initializer='he_normal', padding='same')(p1)
c2 = tf.keras.layers.BatchNormalization()(c2)
p2 = tf.keras.layers.MaxPool2D(pool_size=(3,3),strides=2)(c2)

c3 = tf.keras.layers.Conv2D(filters = 384,kernel_size=(3,3),strides = 1, activation='relu', kernel_initializer='he_normal', padding='same')(p2)
c3 = tf.keras.layers.BatchNormalization()(c3)

c4 = tf.keras.layers.Conv2D(filters = 384,kernel_size=(3,3),strides = 1, activation='relu', kernel_initializer='he_normal', padding='same')(c3)
c4 = tf.keras.layers.BatchNormalization()(c4)

c5 = tf.keras.layers.Conv2D(filters = 256,kernel_size=(3,3),strides = 1, activation='relu', kernel_initializer='he_normal', padding='same')(c4)
c5 = tf.keras.layers.BatchNormalization()(c5)
outputs = tf.keras.layers.MaxPool2D(pool_size=(3,3),strides=2)(c5)

# d1 = tf.keras.layers.Flatten()(p2)

# f1 = tf.keras.layers.Dense(4096,activation='relu')(d1)
# f1 = tf.keras.layers.BatchNormalization()(f1)
# f1 = tf.keras.layers.Dropout(0.1)(f1)

# f2 = tf.keras.layers.Dense(4096,activation='relu')(f1)
# f2 = tf.keras.layers.BatchNormalization()(f2)
# f2 = tf.keras.layers.Dropout(0.1)(f2)

# outputs = tf.keras.layers.Dense(30,activation='softmax')(f2)
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import sparse_categorical_crossentropy

dl_model = tf.keras.Model(inputs=[inputs], outputs=[outputs])
dl_model.compile(optimizer='Adam', loss=tf.keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
```

This function is used to convert the pixel of multiple channels into a 2d array which is converted to dataframe

```python
def binarze(pred):
    bi = len(pred)
    arr = []
    for j in range(0,bi):
        img = pred[j]
        col_a = [None]*9216
        col_c=0
        for i in img:
            for col in i:
                for each in range(0,len(col)):
                    col_a[col_c]=col[each]
                    col_c = col_c+1
        arr.append(col_a)
    return arr
df = binarze(pred)
df2 = binarze(pred2)
```
